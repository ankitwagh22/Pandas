# Removing column from the dataframe

```python
import pandas as pd
df = pd.read_csv("D:\\studentmarks1.csv")
print(df)
```

```python
# adding new column total
df["total"] = df["telugu"]+df["english"]+df["hindi"]+df["maths"]+df["science"]+df["social"]
print(df)
```

```python
print(df)
```

```python
# Removing the total column form dataframe
# Syntax :dataframe.drop(columns="col name")
df.drop(columns="total")
```

```python
print(df)
```

```python
df.drop(columns="telugu")
```

```python
print(df)
```

```python
df.drop(columns=["maths","science"],inplace=True)
```

```python
print(df)
```

```python
print(df)
```

```python
df.drop(columns="total", inplace=True)
```

```python
print(df)
```

```python
# Remove the Column Values Based on Column Number
# Syntax:  dataframe.drop(columns=DataFrameObj.columns[Col Number])
df.drop(columns=df.columns[4],inplace=True)
```

```python
print(df)
```

```python
# Removing Multiple Columns based on Column Numbers
# Syntax: dataframe.drop(columns=DataFrameObj.columns[[Col Number1,Col Number2....]])
df.drop(columns=df.columns[[2,3]],inplace=True)
```

```python
print(df)
```
