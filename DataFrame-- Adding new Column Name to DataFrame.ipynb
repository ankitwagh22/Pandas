# Data Filtering and Updating the values

```python
# Assig grades
# Grading Rules
# "Distinction"----Provided percent >=72
# "First"---------Provded percent ranges between less than 72 and greater than or equal to 65
# "Second"--------provided Percent ranges between less than 65 and greater than or equal to 60
# "Thrid"-------Provided Percent less than 60
```

```python
df["grade"] = None
print(df)
```

```python
df.loc[df["percent"]>=72,["grade"]] = "DISTINCTION"
print(df)
```

```python
# "First"---------Provded percent ranges between less than 72 and greater than or equal to 65
df.loc[(df["percent"]<72) & (df["percent"]>=65),["grade"]] = "FIRST"
print(df)
```

```python
# "Second"--------provided Percent ranges between less than 65 and greater than or equal to 60
df.loc[(df["percent"]<65) & (df["percent"]>=60), ["grade"]] = "SECOND"
print(df)
```

```python
# "Thrid"-------Provided Percent less than 60
df.loc[df["percent"]<60, ["grade"]] = "THIRD"
print(df)
```

```python
# get the students name, percentage and grades
df.loc[::,["name","percent","grade"]]
```

```python
# get the Students name and percentage who passed in Distinction
df.loc[df["grade"] == "DISTINCTION"][["name","percent"]]
```

```python
# get the Students name and percentage who passed in First
df.loc[df["grade"] == "first".upper()][["name","percent","grade"]]
```

```python
# get the Students names and percentage who passed in Second
df.loc[df["grade"] == "Second".upper()][["name","percent"]]
```

```python
# get the Students names and percentage who passed in Third
df.loc[df["grade"] == "Third".upper()][["name","percent"]]
```

```python
# get the Students names and percentages who passed in Distinction and First class
df.loc[(df["grade"] == "DISTINCTION") | (df["grade"] == "FIRST"), ["name","percent","grade"]]
```
