# Sorting the Data of Dataframe 

```python
import pandas as pd
df = pd.read_csv("D:\\studentmarks1.csv")
print(df)
```

```python
# we can add total and percentage column 
df["total"] = df["telugu"]+df["maths"]+df["english"]+df["science"]+df["social"]+df["hindi"]
print(df)
```

```python
df["percentage"] = round((df["total"]/600)*100)
print(df)
```

```python
df.sort_values(["percentage"])
```

```python
print(df)
```

```python
df.sort_values(["percentage"],inplace=True)
```

```python
print(df)
```

```python
df.loc[::,["name","percentage"]]
```

```python
df.sort_values(["total"],ascending=False,inplace=True)
```

```python
print(df)
```

```python
# Top Five students
df.loc[::,["name","total"]].head()
```

```python
# Last five students
df.loc[::,["name","total"]].tail()
```

```python
df.loc[(df["total"]>=380) & (df["total"]<=420),["name","total"]]
```

```python
# get the five topper using loop
for rec in df.loc[::].iterrows():
    print(rec)
    print('--------------------')
```

```python
# get only names of first 3 topper students
df.loc[::,["name"]].head(3)
```

```python
# get the last 3 students names
df.loc[::,["name"]].tail(3)
```

```python
print(df)
```

```python
df.sort_values(["maths"],ascending=False,inplace=True)
print(df)
```

```python
# get the 5 maths topper with their names and marks
df.head()[["name","maths"]]
```

```python
# get the last 5 student in maths with their names and marks
df.tail()[["name","maths"]]
```

```python
# get the top five highest percentage students with names
# first we need to sort the percentage column and then get it
df.sort_values(["percentage"],ascending=False, inplace=True)
```

```python
print(df)
```

```python
# Now the required data
df.head()[["name","percentage"]]
```

```python
# get the last five lowest percentage students with names
df.tail()[["name","percentage"]]
```

```python
# get the names in alphabetical order
df.sort_values(["name"],inplace=True)
```

```python
print(df)
```

```python
# get the names in uppercase letters
for name in df["name"]:
    print(name.upper())
```

```python
print(df)
```

```python
# replace the names of students into upper case letters
df.loc[::,["name"]] = df["name"].str.upper()
```
