# Exporting Data Frame data to CSV File and Excel Files

```python
import pandas as pd
```

```python
df = pd.read_csv("D:\\Full stack Python Notes\\Module-2 Python for Data Science\\Pandas Class Notes on 21st OCT 2024\\studentmarks1.csv")
print(df)
```

```python
# Adding the total marks column
df["total"]=df["telugu"]+df["english"]+df["hindi"]+df["maths"]+df["science"]+df["social"]
print(df)
```

```python
# Adding percentages column
df["percent"] = (df["total"]/600)*100
print(df)
```

```python
df["percent"] = round(df["percent"],2)
print(df)
```

```python
df[df["percent"]>=72]
```

```python
df[(df["percent"]>=60) & (df["percent"]<=70)]
```

```python
df.loc[(df["percent"]>=60) & (df["percent"]<=70)]
```

```python
# assigining the grades based on given values 
# Assig grades
# Grading Rules
# "Distinction"----Provided percent >=72
# "First"---------Provded percent ranges between less than 72 and greater than or equal to 65
# "Second"--------provided Percent ranges between less than 65 and greater than or equal to 60
# "Thrid"-------Provided Percent less than 60
```

```python
df["grade"] = None
print(df)
```

```python
df.loc[df["percent"]>=72,["grade"]] = "DISTINCTION"
print(df)
```

```python
df.loc[(df["percent"]>=65) & (df["percent"]<72),["grade"]] = "FIRST CLASS"
print(df)
```

```python
df.loc[(df["percent"]>=60) & (df["percent"]<65),["grade"]] = "SECOND CLASS"
print(df)
```

```python
df.loc[df["percent"]<60,["grade"]] = "THIRD CLASS"
print(df)
```

```python
# To Export the DataFrame object data to the csv file
# OR Write DataFrame Object data to CSV File
df.to_csv("D:\\Full stack Python Notes\\Module-2 Python for Data Science\\Z-data_samples\\final_student_result.csv")
print("Data inserted into csv file")
```
